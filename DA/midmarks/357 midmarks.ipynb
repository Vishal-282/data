# assigning the grade with A,B,C,D,E¶

```python
def assign_grade(percentage):
    if percentage >= 90:
        return 'A'
    elif percentage >= 80:
        return 'B'
    elif percentage >= 70:
        return 'C'
    elif percentage >= 60:
        return 'D'
    else:
        return 'F'

y['Grade'] = y['percentage'].apply(assign_grade)
y
```
